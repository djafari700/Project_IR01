Preprocessing
1. Load the data (located in the data folder)
2. Perform tokenization
3. Delete stop words
4. According to the data, remove any extra characters (for example: tags, punctuation, etc.) in the
data and clean the data.
5. Perform stemming

Indexing
1. After collecting all the tokens, Build an inverted index

Empty dictionary to display reverse pointer for index & doc counting

```python
indexed_dictionary = {}
doc_count_dictionary = {}
original_list = []
```

function for Preprocessing and Indexing

```python

def preprocessing_function(directory_name):
  # 1-Load the data (located in the data folder)
  file_path = 'drive/MyDrive/Data/' + directory_name
  with open(file_path,'r') as f:
    text = f.read()
  text
  example_sent = text
  directory_name


  # 2-Perform tokenization
  word_tokens = text.split()
  word_tokens


  # 3-Delete stop words
  # nltk.corpus & nltk.tokenize & nltk libraries

  # need stopwords & punkt

  stop_words = set(stopwords.words('english'))

  # converts the words in word_tokens to lower case and then checks whether
  # they are present in stop_words or not
  filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
  filtered_sentence


  # 4.According to the data, remove any extra characters (for example: tags, punctuation, etc.) in the data and clean the data
  # 4.1 Remove the HTML tags

  # rejex library

  pattern = '<[^<]+?>'
  no_tag = [re.sub(pattern, "", w) for w in filtered_sentence]
  no_tag

  # 4.2 Remove trailing / leading special characters from strings list

  res = [x.strip() for x in no_tag]
  res

  # 4.3 Remove punctuation from a List of strings

  # sting library

  new_list = [''.join(char for char in item
                      if char not in string.punctuation)
              for item in res if item != '']

  #delete empty strings from list python
  new_list = [x for x in new_list if x != '']
  new_list


  # 5.Perform stemming
  # nltk.stem
  ps = PorterStemmer()
  # choose some words to be stemmed

  final_list = [ps.stem(w) for w in new_list]
  final_list

  # Nexst step
  # 1.Build dictionaries

  # 1.1Build an inverted index
  for i in final_list:
    indexed_dictionary[i] = indexed_dictionary.get(i, 0) + 1
  indexed_dictionary

  # 1.2Unique repetition in all documents
  final_set  = set(final_list)
  for i in final_set:
    doc_count_dictionary[i] = doc_count_dictionary.get(i, 0) + 1
  doc_count_dictionary
  original_list.append(' '.join(final_list))
```

```python
import numpy as np

documents = original_list

tokenized_docs = [doc for doc in documents]

# Compute the term frequencies
term_frequencies = []
for doc in tokenized_docs:
    freqs = {}
    for word in doc:
        freqs[word] = freqs.get(word, 0) + 1
    term_frequencies.append(freqs)

# Create the document-term matrix
terms = sorted(set().union(*tokenized_docs))
num_docs = len(documents)
num_terms = len(terms)
dt_matrix = np.zeros((num_docs, num_terms))
for i, freqs in enumerate(term_frequencies):
    for j, term in enumerate(terms):
        dt_matrix[i, j] = freqs.get(term, 0)

# Compute the inverse document frequencies
idf = np.log(num_docs / np.count_nonzero(dt_matrix, axis=0))

# Compute the TF-IDF weights
tfidf_matrix = dt_matrix * idf

# Perform k-means clustering
k = 3  # Number of clusters
num_iterations = 100  # Number of iterations
num_docs = tfidf_matrix.shape[0]
num_terms = tfidf_matrix.shape[1]
cluster_centers = tfidf_matrix[np.random.choice(num_docs, k, replace=False)]
for _ in range(num_iterations):
    # Assign documents to clusters
    distances = np.linalg.norm(tfidf_matrix[:, np.newaxis] - cluster_centers, axis=-1)
    labels = np.argmin(distances, axis=1)

    # Update cluster centers
    for cluster_idx in range(k):
        mask = labels == cluster_idx
        cluster_centers[cluster_idx] = np.mean(tfidf_matrix[mask], axis=0)

# Print the results
for i, document in enumerate(documents):
    print(f"Document '{document}' belongs to cluster {labels[i]}")
```

```python

# Perform k-means clustering with L0 norm
k = 2  # Number of clusters
num_iterations = 100  # Number of iterations
num_docs = tfidf_matrix.shape[0]
num_terms = tfidf_matrix.shape[1]
cluster_centers = tfidf_matrix[np.random.choice(num_docs, k, replace=False)]
for _ in range(num_iterations):
    # Assign documents to clusters based on L0 norm
    distances = np.linalg.norm(tfidf_matrix[:, np.newaxis] - cluster_centers, ord=0, axis=-1)
    labels = np.argmin(distances, axis=1)

    # Update cluster centers
    for cluster_idx in range(k):
        mask = labels == cluster_idx
        cluster_centers[cluster_idx] = np.mean(tfidf_matrix[mask], axis=0)

# Print the results
print("Results with L0 norm:")
for i, document in enumerate(documents):
    print(f"Document '{document}' belongs to cluster {labels[i]}")


```
